```python
from nlinec.data.load import load_data
from nlinec.data.preprocessing import get_granularity, get_type, construct_hypothesis
```

```python
data = load_data('augmented_train.json')
data.head()
```

```python
data['granularity'] = data['type'].apply(get_granularity)
data.head()
```

```python
data['type_granularity_2'] = data['type'].apply(lambda type: get_type(type, 2))
data.dropna(inplace=True, subset=['type_granularity_2'])
data.reset_index(drop=True, inplace=True)
data.head()
```

```python
data['hypothesis'] = data.apply(lambda row: construct_hypothesis(row['mention_span'], row['type_granularity_2']), axis=1)
data.head()
```
